## Calculating the Risk of a Security

*Suggested Answers follow (usually there are multiple ways to solve a problem in Python).*

Download the data for Microsoft (‘MSFT’) from Yahoo Finance for the period ‘2000-1-1’ until today. <br />
Assess the daily and the annual risk of ‘MSFT’. Repeat the exercise for Apple for the same timeframe. 

```python
import numpy as np
import pandas as pd
from pandas_datareader import data as wb
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### MSFT

Daily risk:

```python
MSFT=wb.DataReader('MSFT', start='2000-1-1', data_source='yahoo')
```

```python
returns_MSFT=np.log(MSFT['Adj Close']/MSFT['Adj Close'].shift(1))
```

```python
returns_MSFT=returns_MSFT.fillna(value=0)
```

Daily and annual risk:

```python
print('Daily mean returns {:.4f} '.format(returns_MSFT.mean()))
print('Daily risk {:.4f} '.format(returns_MSFT.std()))
```

```python
print('Annual mean returns {:.4f} '.format((returns_MSFT*250).mean()))
print('Annual risk {:.4f} '.format((returns_MSFT*250**0.5).std()))
```

### Apple

Daily risk:

```python
AAPL=wb.DataReader('AAPL', start='2000-1-1', data_source='yahoo')
```

```python
returns_AAPL=np.log(AAPL['Adj Close']/AAPL['Adj Close'].shift(1))
```

```python
returns_AAPL.fillna(value=0)
```

```python
print ('Daily returns AAPL {:.4f}'.format(returns_AAPL.mean()))
print ('Daily risk AAPL {:.4f}'.format(returns_AAPL.std()))
```

Annual risk:

```python
print('Annual mean returns {:.4f} '.format(returns_AAPL.mean()*250))
print('Annual risk {:.4f} '.format(returns_AAPL.std()*250**0.5))
```
